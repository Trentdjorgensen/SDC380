```python
import pandas as pd
from IPython.display import display, Markdown
from datetime import date

# Define student ID and today's date
student_id = "Trejor6927"
todays_date = date.today()

# Display header
display(Markdown(f"### 4.6 Project - {student_id} - {todays_date}"))

# Load dataset
file_path = r"C:/Files/Excel/Prescriptive_PythonR/2019_Happiness_Index_v4.xlsx"

happiness_df = pd.read_excel(file_path)

# Preview dataset
happiness_df.head()
```

```python
# Display header
display(Markdown(f"### 4.6 Project - {student_id} - {todays_date}"))

# Identify categorical columns
categorical_cols = happiness_df.select_dtypes(include=['object']).columns

# Apply one-hot encoding
happiness_df = pd.get_dummies(happiness_df, columns=categorical_cols, drop_first=True)

# Preview transformed dataset
happiness_df.head()
```

```python
from sklearn.model_selection import train_test_split
# Display header
display(Markdown(f"### 4.6 Project - {student_id} - {todays_date}"))

# Define features (X) and target (y)
X = happiness_df[[
    "GDP per capita",
    "Social support",
    "Healthy life expectancy",
    "Freedom to make life choices",
    "Generosity",
    "Perceptions of corruption"
]]

y = happiness_df["Score"]

# 70-30 split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
# Display header
display(Markdown(f"### 4.6 Project - {student_id} - {todays_date}"))

# Initialize model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

print("Model training complete.")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
# Display header
display(Markdown(f"### 4.6 Project - {student_id} - {todays_date}"))

# Predictions
y_pred = model.predict(X_test)

# Evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R²):", r2)
```

```python
# Display header
display(Markdown(f"### 4.6 Project - {student_id} - {todays_date}"))

# Create feature importance DataFrame
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

# Rank by absolute value
feature_importance["Absolute Coefficient"] = abs(feature_importance["Coefficient"])
feature_importance = feature_importance.sort_values(
    by="Absolute Coefficient", ascending=False
)

print("Feature Importance Ranking:")
print(feature_importance[["Feature", "Coefficient"]])
```
